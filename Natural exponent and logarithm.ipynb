```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3, 3,  99)

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=[8, 10])
ax[0].plot(x, np.exp(-np.abs(x)))
ax[0].set_xlabel('x')
ax[0].set_ylabel('$e^x$')
ax[1].plot(x, np.log(x))
ax[1].set_xlabel('x')
ax[1].set_ylabel('log(x)')
```

```python
# Exercise

x = np.linspace(.0001,10, 10)
y1 = np.log(np.exp(x))
y2 = np.exp(np.log(x))

plt.plot(x, y1, label='$log(e^x)$')
plt.plot(x, y2, 'ro', label='$e^log(x)$')
plt.legend()
```
