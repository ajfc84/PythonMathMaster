```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Exercise

x = np.array([-5, 5])
m = -.3
b = .1

y = m*x + b

plt.plot(x, y, label='y=%s%s+%s' % (m, x, b))
plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0, 0], 'k--')
plt.plot([0, 0], axis.get_ylim(), 'k--')

plt.legend()
```
