```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 260)

angle = np.pi/4
plt.plot([0, np.cos(angle)], [0, np.sin(angle)])

plt.plot(np.cos(x), np.sin(x))
plt.axis('square')
plt.xlabel('cos(x)')
plt.ylabel('sin(x)')
plt.plot([-2, 2], [0, 0], 'k')
plt.plot( [0, 0], [-2, 2], 'k')
```
