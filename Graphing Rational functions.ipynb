```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Math, display
import sympy as sym
```

```python
x = range(-3, 4)
y = np.zeros(len(x))


for i in range(0, len(x)):
    y[i] = -x[i]**2 + 2
plt.plot(x, y, 's-')
```

```python
x = np.linspace(-3, 4, 144)
y = -np.sqrt(abs(x)) + 2

plt.plot(x, y, 'm-')
```
