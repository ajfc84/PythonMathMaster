```python
import matplotlib.pyplot as plt
```

```python
x = 3
y = 5
plt.plot(x, y, 'ro')

plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.grid()
```

```python
x = [-4, 5, 2, 4, 5, 6, 3]
y = [0, 3, 4, 5, 2, -3, 4]

plt.plot(x, y, 'o')
plt.axis('square')
axis = plt.gca()
axis.set_ylim(0, 7)
```
