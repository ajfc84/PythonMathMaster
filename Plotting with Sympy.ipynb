```python
import sympy as sym
import sympy.plotting as symplot
```

```python
from sympy.abc import x

y = x**2

p = symplot.plot(y, show=False)
p.xlim = [0, 50]
p[0].line_color = 'm'
p.title = 'This is a nice plot'
p.show()
```
