```python
import matplotlib.pyplot as plt
import math
```

```python
x = [1, 5]
y = [1, 4]

plt.plot(x, y, 'r')
ax = plt.gca()
ax.set_xlim(0, 6)
ax.set_ylim(0, 5)
```

```python
x = [0, 5]
y = [0, 4]

plt.plot(x, y, 'r')
plt.plot(x, y, 'ro')
plt.axis('square')
plt.axis([-6, 6, -6, 6])
plt.grid()

plt.plot([-6, 6], [0,0], 'k')
plt.plot([0, 0], [-6, 6], 'k')
```

```python
# Exercise

X = range(-20, 20 +1)
f = lambda x : math.sqrt(abs(x))
```

```python
for x in X:
    plt.plot([0, x], [0, f(x)])
```
